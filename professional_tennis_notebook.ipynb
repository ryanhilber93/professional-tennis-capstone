```python
#calculating total aces from winner and loser aces
all_matches['total_ace'] = all_matches['w_ace'] + all_matches['l_ace']
```

```python
all_matches['total_points'] = all_matches['w_svpt'] + all_matches['l_svpt']
all_matches['total_1stIn'] = all_matches['w_1stIn'] + all_matches['l_1stIn']
all_matches['total_2ndserve'] = all_matches['total_points'] - all_matches['total_1stIn']
all_matches['total_1stWon'] = all_matches['w_1stWon'] + all_matches['l_1stWon']
all_matches['total_2ndWon'] = all_matches['w_2ndWon'] + all_matches['l_2ndWon']
all_matches['total_1st_perc'] = 100 * all_matches['total_1stIn'] / all_matches['total_points']
all_matches['total_1st_win_perc'] = 100 * all_matches['total_1stWon'] / all_matches['total_1stIn']
all_matches['total_2nd_win_perc'] = 100 * all_matches['total_2ndWon'] / all_matches['total_2ndserve']
all_matches['aces_per_point'] = all_matches['total_ace'] / all_matches['total_points']
all_matches['df_per_point'] = all_matches['total_df'] / all_matches['total_points']
all_matches.tail()
```

```python
#again only using 1990s and later
m_winners90 = m_final_90['winner_name'].value_counts()
m_winners90 = m_winners90.to_frame()
m_winners90.reset_index(inplace=True)
m_winners90.columns = ['name', 'masters']
m_winners90
```

```python
#calculating total wins of grand slam tournaments by player
gs_winners = gs_final['winner_name'].value_counts()
gs_winners = gs_winners.to_frame()
gs_winners.reset_index(inplace=True)
gs_winners.columns = ['name', 'grand_slams']
gs_winners
```

```python
#calculating most recent masters victory by player
m_winners_recent90 = m_final_90.groupby('winner_name').max('year')
m_winners_recent90 = m_winners_recent90[['year']]
m_winners_recent90.reset_index(inplace=True)
m_winners_recent90.columns = ['name', 'most_recent_m']
m_winners_recent90
```

```python
# calculating most recent grand slam victory by player
gs_winners_recent = gs_final.groupby('winner_name').max('year')
gs_winners_recent = gs_winners_recent[['year']]
gs_winners_recent.reset_index(inplace=True)
gs_winners_recent.columns = ['name', 'most_recent_gs']
gs_winners_recent
```

```python
#merging most recent masters and total masters
m_winners90 = m_winners90.merge(m_winners_recent90, on='name', how='left')
m_winners90
```

```python
#merging the result back to original masters table
m_final_90 = m_final_90.merge(m_winners90, left_on='winner_name', right_on='name', how='left')
m_final_90
```
